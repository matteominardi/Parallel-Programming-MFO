```python
x = [10, 25, 50, 100, 200]
y_2 = [12.58, 24.94, 51.29, 99.65, 206.88]
y_4 = [14.04, 17.34, 30.32, 57.47, 117.69]

x_8 = [10, 25, 50, 100, 200, 400, 800]
y_8 = [10.31, 10.26, 17.11, 34.21, 61.97]

y_f = [3.22, 3.43, 6.58, 13.78, 27.01, 37.34, 77.44]

s_y = [11.43, 28.26, 55.34, 109.20, 220.47]

plt.plot(x, s_y, label="Sequential: 1 process", marker='o')
plt.plot(x, y_2, marker='o', label="First Hybrid: 2 processes")
plt.plot(x, y_4, marker='o', label="First Hybrid: 4 processes")
plt.plot(x, y_8, marker='o', label="First Hybrid: 8 processes")
plt.plot(x, y_f[:-2], marker='o', label="Fastest Config First Hybrid: 8 processes")
plt.xlabel('Population Size')
plt.ylabel('Execution Time')
plt.title('First Hybrid MFO Algorithm: Execution Time')
plt.legend()
plt.show()
```
